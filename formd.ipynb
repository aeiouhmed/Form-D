```python
try:
    excel_file = "BFYGA1P1_QQ4V.XLS"
    # Using xlrd engine for .xls files
    df = pd.read_excel(excel_file, sheet_name=None, engine='xlrd')
    print("Sheet names in the Excel file:")
    print(df.keys())
except FileNotFoundError:
    print(f"Error: The file '{excel_file}' was not found. Make sure the file exists in the same directory as this notebook.")
except Exception as e:
    print(f"An error occurred: {str(e)}")
    print("\nNote: For .XLS files (older Excel format), you need to install the 'xlrd' package.")
    print("You can install it by running: pip install xlrd")
```

```python
# Access Sheet 1 from the dictionary of dataframes
sheet1_df = df['Sheet 1']

# Display basic information about the dataset
print("Dataset Info:")
print("-" * 50)
sheet1_df.info()

print("\nFirst few rows of the dataset:")
print("-" * 50)
print(sheet1_df.head())

print("\nBasic statistics of numerical columns:")
print("-" * 50)
print(sheet1_df.describe())
```

```python
# Create a new dataframe with only FORM-D entries
try:
    # Check if the column exists (we'll make this case-insensitive)
    form_flag_col = next(col for col in sheet1_df.columns if 'form' in col.lower() and 'flag' in col.lower())
    
    # Filter the dataframe to get only FORM-D entries
    form_d_df = sheet1_df[sheet1_df[form_flag_col].str.upper() == 'FORM-D']
    
    print(f"Number of FORM-D entries found: {len(form_d_df)}")
    print("\nFirst few rows of FORM-D entries:")
    print("-" * 50)
    print(form_d_df.head())
    
    # Display the columns in the filtered dataframe
    print("\nColumns in the filtered dataframe:")
    print("-" * 50)
    print(form_d_df.columns.tolist())
    
except StopIteration:
    print("Could not find a column related to form flag. Available columns are:")
    print(sheet1_df.columns.tolist())
except Exception as e:
    print(f"An error occurred: {str(e)}")
```
